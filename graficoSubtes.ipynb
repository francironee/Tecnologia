GRAFICO VIAJES SUBTES

```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df1 = pd.read_csv("data/dataset_viajes_sube.csv")
df1 = df1.loc[0:100]
x_values = df1['TIPO_TRANSPORTE'].unique()
y_values = df1['TIPO_TRANSPORTE'].value_counts().tolist()
y_values.reverse()

plt1.bar(x_values,y_values)
plt1.title('Tipo de transporte utilizado por dia')
plt1.xlabel('Tipo de transporte')
plt1.ylabel('Dia')
plt1.show()

plt2.plot(x_values,y_values)
plt2.title('Tipo de transporte utilizado por dia')
plt2.xlabel('Tipo de transporte')
plt2.ylabel('Dia')
plt2.show()
```
